```python
def get_different_datasets(df, n_features):
    columns = np.array(df.drop(['Machine failure'], axis = 1).columns)
    df_ = df[shuffle(columns)[:n_features]]
    features_to_add = np.array(df_.columns)[:3]
    df_=pd.concat([df_, pd.Series(df_[features_to_add[0]]*df_[features_to_add[1]],name=features_to_add[0]+'*'+features_to_add[1]) ], axis=1)
    df_=pd.concat([df_, pd.Series(df_[features_to_add[0]]-df_[features_to_add[2]],name=features_to_add[0]+'-'+features_to_add[2]) ], axis=1)

    return df_
```

```python
def normalize_(arr, arr_n):
    return (arr-np.min(arr_n))/(np.max(arr_n)-np.min(arr_n))
```

```python
for i in np.array(X_test.columns):
    if not i in  ['Air temperature [K]', 'Process temperature [K]']:
        X_test[i] = normalize_(X_test[i], X_train[i])
```

```python
class ClassifingModelV0(nn.Module):
    def __init__(self):
        super().__init__()
        hidden = 20
        self.layer_stack = nn.Sequential(
            nn.Linear(588, hidden),
            nn.ReLU(),
            # nn.LayerNorm(40),
            nn.Linear(hidden, hidden),
            nn.ReLU(),
            # nn.LayerNorm(40),
            # nn.Linear(100, 100),
            # nn.ReLU(),
            nn.Linear(hidden, 1)
        )

    def forward(self, x):
        return self.layer_stack(x)
```

```python
model_0 = ClassifingModelV0().to(device)
```

```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model_0.parameters(), lr=0.0001)
```

```python
epochs = 10000
max_ar = 0
for epoch in range(epochs):

    model_0.train()
    y_logits = model_0(X_train_T).squeeze()
    # y_pred = torch.sigmoid(y_logits)
    # auc_roc = roc_auc_score(y_train_T.detach().numpy(), y_pred)
    loss = loss_fn(y_logits, y_train_T)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    if epoch % 1000 == 0:
        model_0.eval()
        with torch.inference_mode():
            y_logits_test = model_0(X_test_T).squeeze()
            y_pred_test = torch.sigmoid(y_logits_test)
            loss_test = loss_fn(y_logits_test, y_test_T)
            auc_roc_test = roc_auc_score(y_test_T.cpu(), y_pred_test.cpu())
            # if(max_ar<auc_roc_test):
            #     max_ar=auc_roc_test
        print(f'Epoch: {epoch} Loss: {loss:.2f} Test loss: {loss_test:.2f} Test auc roc: {auc_roc_test:.5f}')
```
